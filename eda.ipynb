```python
# Fill numeric columns with their mean
numeric_columns = data.select_dtypes(include=['number'])
data[numeric_columns.columns] = numeric_columns.fillna(numeric_columns.mean())

# Fill non-numeric (categorical) columns with 'Unknown'
non_numeric_columns = data.select_dtypes(exclude=['number'])
data[non_numeric_columns.columns] = non_numeric_columns.fillna('Unknown')

# For a specific column like 'wage', impute using median values grouped by 'industry'
data['wage'] = data.groupby('industry')['wage'].transform(lambda x: x.fillna(x.median()))


# Encoding categorical variables
data_encoded = pd.get_dummies(data, drop_first=True)
from sklearn.impute import KNNImputer
from sklearn.impute import KNNImputer
import pandas as pd

# Separate the numerical and categorical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
categorical_columns = data.select_dtypes(exclude=['float64', 'int64'])

# Use KNN imputer for numerical columns
knn_imputer = KNNImputer(n_neighbors=5)
numeric_data_imputed = pd.DataFrame(knn_imputer.fit_transform(numeric_columns), columns=numeric_columns.columns)

# Handle missing values for categorical columns (using 'mode' or a different method)
categorical_data_imputed = categorical_columns.apply(lambda col: col.fillna(col.mode()[0]))

# Concatenate the imputed numerical and categorical data
data_imputed = pd.concat([numeric_data_imputed, categorical_data_imputed], axis=1)

# Now 'data_imputed' contains the imputed values for both numeric and categorical columns
print(data_imputed.head())




from sklearn.experimental import enable_iterative_imputer  # Enable IterativeImputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
import pandas as pd


data=pd.read_csv("C:/Users/ankur/OneDrive/Desktop/Males.csv")
# Step 1: Identify categorical columns and encode them
categorical_columns = data.select_dtypes(include=['object']).columns

# Use LabelEncoder to encode categorical columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))  # Convert 'no'/'yes' to numbers (0, 1)
    label_encoders[col] = le  # Store the encoder for future use

# Step 2: Apply Iterative Imputer for missing numeric values
mice_imputer = IterativeImputer(max_iter=10, random_state=0)
data_imputed = mice_imputer.fit_transform(data)

# Convert the imputed data back to a DataFrame
data_imputed_df = pd.DataFrame(data_imputed, columns=data.columns)

# Step 3: If needed, decode the categorical columns back to their original values
for col, le in label_encoders.items():
    data_imputed_df[col] = le.inverse_transform(data_imputed_df[col].astype(int))

# Print the imputed data
print(data_imputed_df.head())



# Outlier detection and handling
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[col] = data[col].clip(lower_bound, upper_bound)

# Print the first few rows of the data after outlier handling
print(data.head())


# Scaling numerical data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data.select_dtypes(include=['float64', 'int64'])), columns=data.select_dtypes(include=['float64', 'int64']).columns)

# Splitting dataset into training and testing sets
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Elbow method to find optimal k
inertia = []
k_range = range(1, 11)  # Try k from 1 to 10 clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_encoded.select_dtypes(include=['float64', 'int64']))
    inertia.append(kmeans.inertia_)

# Plotting the Elbow graph
plt.plot(k_range, inertia, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.show()
```
